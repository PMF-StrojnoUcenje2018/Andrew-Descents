```python
import torch
import torchvision
from PIL import Image
import csv
import os
from PIL import ImageFile
ImageFile.LOAD_TRUNCATED_IMAGES = True

classes = ['Albertsons',
 'BJs',
 'CVSPharmacy',
 'Costco',
 'FredMeyer',
 'Frys',
 'HEB',
 'HarrisTeeter',
 'HyVee',
 'JewelOsco',
 'KingSoopers',
 'Kroger',
 'Meijer',
 'Publix',
 'Safeway',
 'SamsClub',
 'ShopRite',
 'Smiths',
 'StopShop',
 'Target',
 'Walgreens',
 'Walmart',
 'Wegmans',
 'WholeFoodsMarket',
 'WinCoFoods']

model = torchvision.models.resnet18()
num_in_features = model.fc.in_features
model.fc = torch.nn.Linear(num_in_features, 25)
model.load_state_dict(torch.load('buto_punoepoha_CPU.pt'))

transform = torchvision.transforms.Compose([
        torchvision.transforms.Resize((224,224)),
        torchvision.transforms.ToTensor(),
        torchvision.transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
        ])

data_dir = 'D:\\Downloads\\mozgalo\\val'
test_dir = 'D:\\Downloads\\mozgalo_test\\robust_ml_challenge_final_testset'


transformed_image = torch.randn(3,224,224)
input_tensor = torch.randn(1,3,224,224)

#treshold = [11.5, 12, 10, 10, 13, 10, 10, 10, 10, 16, 8, 12, 8.9,  10, 10, 10, 10, 10, 10, 9, 10, 8, 10, 12, 12]
#treshold = [11.6, 12.78, 10, 10, 13, 10, 10, 10, 10.7, 16.5, 8, 17.3, 9.7,  10, 10, 10, 10, 11.3, 10, 9, 10.4, 8, 10, 14.1, 12]
treshold = [11.6, 13.3, 10, 10, 13, 10, 10, 10, 10.7, 16.5, 8, 17.3, 10.7,  10, 10, 10, 10, 11.3, 10, 9, 10.4, 8, 10, 14.1, 13.5]

model.eval()
with open('najkonacnije_rjesenje_kopija2.csv','w', newline = '') as csvfile:
    writer = csv.writer(csvfile)
    for index in range(10000):
        if index % 25 == 0:
            print(index)
        image = Image.open(os.path.join(test_dir, str(index)+'.jpg'))
        transformed_image = transform(image)
        input_tensor[0,:,:,:] = transformed_image
        output_tensor = model(input_tensor)
        #print(output_tensor)
        dominant, prediction = torch.max(output_tensor, dim = 1)
        #print(dominant, classes[prediction])
        if float(dominant) > treshold[prediction]:
            writer.writerow([classes[prediction]])
        else:
            writer.writerow(['Other'])
```
